```python
#calculate the number of uppercase & lowercase alphabets in given string.
#string: She shell Seashells by the Sentence
def calc_string(x):
  lower=0
  upper=0
  for i in x:
    if i.isupper()==True:
      upper +=1
    else:
      lower +=1
  return lower,upper
print(calc_string("She shell Seashells by the Sentence"))
```

```python
# write a function that will give sum of numbers in the list

def sum_of_list(lst):
    total = 0
    for num in lst:
        total += num
    return total
print(sum_of_list([3,4,5]))
```

```python
#write the function that will give count of vowels present in the given string
def count_vowels(x):
  count=0
  vowels=['a','e','i','o','u']
  for char in x:
    if char in vowels:
      count += 1
  return count
print(count_vowels("pavani"))
```
